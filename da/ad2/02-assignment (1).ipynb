#### Декораторы
**Декораторы** — это специальный механизм в Python, который позволяет дополнять поведение функции без изменения ее кода. Они «оборачивают» функцию в другую, добавляя нужный функционал — например, логирование, проверку прав, кеширование или измерение времени выполнения.

Синтаксически декоратор — это просто функция, которая принимает другую функцию и возвращает новую, измененную. Обычно декораторы применяются с помощью символа @ над определением функции, например:
```python
@decorator_name
def my_function():
    ...
```
Мы хотим **сравнить время выполнения функций**, реализованных на чистом Python и с использованием NumPy. Конечно, можно было бы вставить код замера времени прямо внутри каждой функции, но в этом случае код стал бы громоздким и повторяющимся, а логика замера времени смешалась бы с основной задачей.

Чтобы сделать код **более аккуратным, читаемым и удобным для повторного использования**, мы воспользуемся **декоратором**.  


#### Задание 1.1. Замер скорости
1. Реализуй декоратор `@benchmark`, который:
    * принимает функцию,
    * измеряет время ее выполнения,
    * выводит результат в **секундах**.  
2. Протестируй работу декоратора на функции `python_mean`, которая вычисляет **среднее арифметическое** элементов списка на чистом Python.
3. В качестве теста передай в функцию список из **1 000 000 элементов**.

````python
import time

def benchmark(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        res = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f'work_time {execution_time}')
        return res
    return wrapper
````

````python
@benchmark
def python_mean(array):
    s = sum(array)
    return s / len(array)

arr = [i for i in range(1_000_000)]
result = python_mean(arr)
print(result)
````

## Головоломка судоку

**NumPy** — это мощный инструмент для работы с массивами и матрицами, который широко используется в анализе данных и научных вычислениях. Чтобы почувствовать себя увереннее при работе с матрицами, мы попробуем реализовать **решатель головоломок судоку**.

### Задание 2.1. Запрос по API

* Напиши функцию `get_sudoku(level)`, которая принимает уровень сложности (`'easy'`, `'medium'`, `'hard'`).
* Функция должна возвращать две матрицы `numpy` размером 9×9: одну — с головоломкой, другую — с полным решением.
* После запроса выведи обе матрицы, чтобы убедиться, что данные получены и преобразованы правильно.

````python
import requests


def get_sudoku(level):

  
    body = {
    "difficulty": level, 
    "solution": True, 
    "array": False 
    }

    headers =  {"Content-Type":"application/json"}

    response = requests.post("https://youdosudoku.com/api/", json=body, headers=headers)

    data = response.json()

    puzzle = np.array([int(i) for i in data['puzzle']])
    puzzle.resize(9, 9)

    solution = np.array([int(i) for i in data['solution']])
    solution.resize(9, 9)

    return puzzle, solution
    

puz, sol = get_sudoku('hard')
print(f'puzzle: {puz}')
print(f'solution: {sol}')

````

````python
import ipytest

ipytest.autoconfig()


def test_solution_in_row(sol):
    '''Проверка в рядах'''
    for el in sol:
        if not np.array_equal(np.sort(el), np.arange(1,10)):
            assert False, f"Некорректная строка: {el}"

def test_solution_in_colums(sol):
    '''Проверка в колонках'''
    for idx in range(9):
        a = np.array([])
        for row in sol:
            a = np.append(a,row[idx])
        if len(np.unique(a)) != 9:
            assert False, f"Некорректная стоблец: {idx}"


def test_solution_in_cub(sol):
    '''Проверка в квадратах'''
    a1 = sol[:3]
    a2 = sol[3:6]
    a3 = sol[6:]
    
    for i in [a1, a2, a3]:
        
        a1_ = np.resize(i[0], [3, 3, 3])
        a2_ = np.resize(i[1], [3, 3, 3])
        a3_ = np.resize(i[2], [3, 3, 3])
        
        cub1 = np.concatenate((a1_[0], a2_[0], a3_[0]))
        cub2 = np.concatenate((a1_[1], a2_[1], a3_[1]))
        cub3 = np.concatenate((a1_[2], a2_[2], a3_[2]))
        
        c1 = np.unique(cub1)
        c2 = np.unique(cub2)
        c3 = np.unique(cub3)
        
        if not(len(c1) == len(c2) == len(c3) == 9):
            assert False
        

test_solution_in_row(sol)
test_solution_in_colums(sol)
test_solution_in_cub(sol)


````

````python
def check_row(arr, idx):
    a = np.unique(arr[idx])
    if len(a) < len(arr[idx]):
        return False
    return True


def check_col(arr, idx):
    arr = np.transpose(arr)
    a = np.unique(arr[idx])
    if len(a) < len(arr[idx]):
        return False
    return True


def check_cub(arr, idx1, idx2):
        arr = arr[idx1:idx2]
        a1_ = np.resize(arr[0], [3, 3, 3])
        a2_ = np.resize(arr[1], [3, 3, 3])
        a3_ = np.resize(arr[2], [3, 3, 3])
        
        cub1 = np.concatenate((a1_[0], a2_[0], a3_[0]))
        cub2 = np.concatenate((a1_[1], a2_[1], a3_[1]))
        cub3 = np.concatenate((a1_[2], a2_[2], a3_[2]))
        
        c1 = np.unique(cub1)
        c2 = np.unique(cub2)
        c3 = np.unique(cub3)
        
        if not(len(c1) == len(c2) == len(c3) == 9):
            return False
        
        return True

def main(board):
  
    grid = np.array(board, dtype=int).copy()
    
    def is_valid(grid, row, col, num):
      
        if num in grid[row]:
            return False
       
        if num in grid[:, col]:
            return False
       
        start_row, start_col = 3 * (row // 3), 3 * (col // 3)
        block = grid[start_row:start_row+3, start_col:start_col+3]
        if num in block:
            return False
        return True

    def solve(grid):
        for i in range(9):
            for j in range(9):
                if grid[i][j] == 0: 
                    for num in range(1, 10):
                        if is_valid(grid, i, j, num):
                            grid[i][j] = num
                            if solve(grid):
                                return True
                            grid[i][j] = 0 
                    return False
        return True  

    if solve(grid):
        return grid
    else:
        return None 
    


save = []
for level in ['hard', 'medium','easy']:
    for _ in range(5):
        t, sol = get_sudoku(level)
        print("=========")
        a = main(t)
        print(f'answer {a}')
        print(f'puzzle {t}')
        save.append(a)


for i in save:
    test_solution_in_row(i)
    test_solution_in_colums(i)
    test_solution_in_cub(i)

````

## Сингулярное разложение (SVD)

В библиотеке NumPy реализовано множество полезных математических функций: операции с массивами, линейная алгебра, вычисление статистических показателей, тригонометрические и экспоненциальные функции, преобразования Фурье, решение систем уравнений, а также **сингулярное разложение матриц**.

Любую матрицу (A) можно разложить следующим образом:

$$A = U \times S \times V^T$$

Где:
* $U$ — матрица **левых сингулярных векторов**;
* $S$ — **диагональная матрица сингулярных чисел**;
* $V^T$ — матрица **правых сингулярных векторов** (транспонированная).

Это разложение называется **сингулярное разложение**, и оно широко используется в **анализе данных** и **машинном обучении**:

* **сжатие изображений** и сигналов;
* **рекомендательные системы** (например, в коллаборативной фильтрации);
* **выделение основных компонентов** (PCA);
* **аппроксимация матриц** и уменьшение размерности;
* в задачах **анализ текста** (Latent Semantic Analysis).

Матрица $S$ содержит **сингулярные числа**, которые несут основную информацию о матрице (например, изображении).

Если сохранить только **k наибольших сингулярных чисел**, мы можем получить приближенную матрицу:

$$A_k = U[:, :k] \times S[:k, :k] \times V^T[:k, :]$$

* Это позволяет **сжимать изображения**, сохраняя ключевую информацию и убирая шум.
* Чем меньше (k), тем сильнее сжатие, но при этом может теряться часть деталей.

В этом проекте мы воспользуемся SVD для сжатия изображений, чтобы на практике увидеть, как сохраняются важные признаки картинки при уменьшении ее размерности.

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


image = Image.open("C://Users/User/Desktop/school 21/AI_Data_Analytics.Project_2.ID_1577600-2/misc/jackal.png")
image_array = np.array(image)


is_color = len(image_array.shape) == 3


if is_color:
    U_list, S_list, V_list = [], [], []
    for i in range(3):
        channel_data = image_array[:, :, i].astype(float)
        U, S, Vt = np.linalg.svd(channel_data, full_matrices=False)
       
        U_list.append(U)
        S_list.append(S)
        V_list.append(Vt)  
    original = image_array
else:
    gray_array = image_array.astype(float)
    U, S, Vt = np.linalg.svd(gray_array, full_matrices=False)
    V = Vt  
    original = gray_array


k_values = [100, 50, 25, 20, 10]


fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()


if is_color:
    axes[0].imshow(original.astype(np.uint8))
else:
    axes[0].imshow(original, cmap='gray')
axes[0].set_title('Оригинальное изображение')
axes[0].axis('off')


for idx, k in enumerate(k_values, 1):
    if is_color:
        compressed_channels = []
        for i in range(3):
            U_k = U_list[i][:, :k]              
            S_k = np.diag(S_list[i][:k])         
            Vt_k = V_list[i][:k, :]              

            A_k = U_k @ S_k @ Vt_k               
            compressed_channels.append(A_k)

        compressed_image = np.stack(compressed_channels, axis=2)
        compressed_image = np.clip(compressed_image, 0, 255).astype(np.uint8)
    else:
        U_k = U[:, :k]           
        S_k = np.diag(S[:k])     
        Vt_k = V[:k, :]          

        compressed_image = U_k @ S_k @ Vt_k
        compressed_image = np.clip(compressed_image, 0, 255)

    
    ax = axes[idx]
    if is_color:
        ax.imshow(compressed_image)
    else:
        ax.imshow(compressed_image, cmap='gray')
    ax.set_title(f'k = {k}')
    ax.axis('off')

   
    m, n = (image_array.shape[0], image_array.shape[1]) if is_color else image_array.shape
    if is_color:
        original_size = 3 * m * n
        compressed_size = 3 * (m * k + k + k * n)
    else:
        original_size = m * n
        compressed_size = m * k + k + k * n

    compression_ratio = compressed_size / original_size
    print(f"k = {k}: степень сжатия = {compression_ratio:.3f}")

plt.suptitle('Сжатие изображения с помощью SVD', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()


fig2, axes2 = plt.subplots(1, len(k_values), figsize=(15, 4))
for i, k in enumerate(k_values):
    if is_color:
        compressed_channels = []
        for j in range(3):
            U_k = U_list[j][:, :k]
            S_k = np.diag(S_list[j][:k])
            Vt_k = V_list[j][:k, :]
            A_k = U_k @ S_k @ Vt_k
            compressed_channels.append(A_k)
        compressed_image = np.stack(compressed_channels, axis=2)
        compressed_image = np.clip(compressed_image, 0, 255).astype(np.uint8)
        axes2[i].imshow(compressed_image)
    else:
        U_k = U[:, :k]
        S_k = np.diag(S[:k])
        Vt_k = V[:k, :]
        compressed_image = U_k @ S_k @ Vt_k
        compressed_image = np.clip(compressed_image, 0, 255)
        axes2[i].imshow(compressed_image, cmap='gray')
    axes2[i].set_title(f'k = {k}')
    axes2[i].axis('off')

plt.suptitle('Влияние k на качество изображения', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()


print("\n" + "="*50)
print("ИНФОРМАЦИЯ О СЖАТИИ:")
print("="*50)
print(f"Оригинальное изображение: {image_array.shape}")
total_singular = S_list[0].shape[0] if is_color else S.shape[0]
print(f"Всего сингулярных чисел: {total_singular}")

for k in k_values:
    m, n = (image_array.shape[0], image_array.shape[1]) if is_color else image_array.shape
    if is_color:
        orig_size = 3 * m * n
        comp_size = 3 * (m * k + k + k * n)
    else:
        orig_size = m * n
        comp_size = m * k + k + k * n

    compression_percent = (1 - comp_size / orig_size) * 100
    ratio = orig_size / comp_size
    retained_pct = (k / total_singular) * 100

    print(f"\nk = {k}: сохранено {k}/{total_singular} сингулярных чисел ({retained_pct:.1f}%)")
    print(f"  Коэффициент сжатия: {ratio:.2f}:1")
    print(f"  Экономия памяти: {compression_percent:.1f}%")
````
