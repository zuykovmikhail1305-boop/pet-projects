## 1. Инвертированный индекс

Информационный поиск (Information Retrieval) — это область компьютерных наук, которая изучает методы поиска нужной
информации в больших коллекциях данных. Классический пример — поисковые системы в интернете, которые позволяют найти 
документы, статьи или веб-страницы по ключевым словам.

Цель информационного поиска — не просто найти документы, содержащие слово, но и оценить их релевантность запросу 
пользователя. Одним из основных инструментов в информационном поиске является **инвертированный индекс**.

**Инвертированный индекс** — это структура данных, которая для каждого уникального слова хранит список документов или 
частей текста, где это слово встречается. Проще говоря, он «переворачивает» данные: вместо списка документов с их 
словами создается список слов с привязкой к документам.

### Задание 1.3. Построение инвертированного индекса
Реализуй **инвертированный индекс** для поиска слов в предложениях с помощью класса `InvertedIndex`.

**Что нужно сделать:**

**Протестируй работу индекса:**

1. Построй индекс на списке предложений книги.
2. **Сохрани** индекс в файл.
3. **Загрузи** индекс из файла в **новый объект**.
4. Выведи количество **уникальных слов** в загруженном индексе.

```python
from typing import List, Dict
from collections import defaultdict
import pickle
```

```python
class InvertedIndex:


    def __init__(self):
        self.word2doc = defaultdict(list)

    def build_index(self, documents: List[List[str]]):
       
        for idx, sen in  enumerate(documents):
            for j in sen:
                self.word2doc[j.lower()].append(idx)


    def save_index(self, filepath: str):

        with open(filepath, 'wb') as f:
            pickle.dump(self.word2doc, f)

    def load_index(self, filepath: str):

        with open(filepath, 'rb') as f:
            loaded_data = pickle.load(f)
            self.word2doc = loaded_data

    def __len__(self):
        return len(self.word2doc)

                
```

```python
II = InvertedIndex()
print(II.word2doc['война'])
II.build_index(word)

II.save_index("pik1.pkl")
print(II.word2doc['война'])

II.load_index("pik1.pkl")
print(II.word2doc['война'])

II.__len__()
```

### Задание 1.4. Поиск по индексу
* Реализуй метод `query` класса `InvertedIndex`, который выполняет поиск слов в индексе.
* Метод принимает **список слов** и возвращает список id предложений, в которых встречаются **все указанные слова** (операция «и»).
* Результат должен быть **отсортирован по возрастанию** id предложения.
* Проверь работу метода `query` на следующих примерах:
  * слово: `университет`;
  * слова: `война` и `мир`;
  * слово: `школа`.

```python
def query(self, words: List[str]) -> List[int]:

    if not words:
        return []
    

    result_set = set(self.word2doc.get(words[0], []))
    
    for word in words[1:]:
        word_docs = set(self.word2doc.get(word, []))
        result_set = result_set.intersection(word_docs)
    
    return sorted(list(result_set))

InvertedIndex.query = query
```

## Бонусное задание. TF-IDF

Построив **инвертированный индекс**, мы научились быстро находить документы, содержащие определенные слова. Однако этого недостаточно для реального поиска. Представь, что ты ищешь слово *«мир»* в романе *«Война и мир»*. Индекс вернет сотни предложений, но не все они будут одинаково полезны.

В поисковых системах важно не только найти документы, где встречается слово, но и **оценить, насколько это слово важно именно в данном документе**. Для этого используется одна из ключевых идей информационного поиска — **TF-IDF (Term Frequency — Inverse Document Frequency)**.

* **TF** показывает, насколько часто слово встречается в документе.
* **IDF** снижает вес слов, которые встречаются слишком часто в коллекции (например, «и», «он», "the").
* В итоге TF-IDF позволяет отличать «шумовые» слова от действительно значимых.

```python
class Tfidfindex(InvertedIndex):
    def __init__(self):
        super().__init__()

    def _tf(self, document: List[str]) -> Dict[str, float]:  
        """
        Расчет term frequency (TF) для одного документа
        document: список слов в документе
        return: словарь {слово: tf_значение}
        """
        tf_dict = dict()
        n = len(document)
        for word in set(document):
            if checking(word):
                tf_dict[word] = 0.000001
            else:  
                tf_dict[word] = document.count(word) / n

        return tf_dict

        
```

### Задание 2.3. Реализация IDF
* В классе `TfidfIndex` реализуй метод `_idf`.
* Выведи топ-5 слов с самым маленьким значением.

```python
import math
def _idf(self, documents: List[List[str]]) -> Dict[str, float]:
    idf_dict = dict()
    n1 = len(documents)
    for i in self.word2doc.keys(): 
        idf_dict[i] = math.log(n1 / len(self.word2doc[i]))
    
    sorted_dict = dict(sorted(idf_dict.items(), key=lambda item: item[1]))
    return sorted_dict

        
Tfidfindex._idf = _idf
```

### Задание 2.4. Реализация TF-IDF
* В классе `TfidfIndex` реализуй метод `tfidf`, который принимает список документов и возвращает матрицу TF-IDF.
* Рассчитай TF-IDF для всей книги.
* Выведи полученные значения для слов `война` и `мир` для первого предложения.

```python
def compute_tfidf(self, documents: List[List[str]]) -> Dict[int, Dict[str, float]]:
    """
    Вычисление TF-IDF для всех слов во всех документах
    documents: список документов, каждый документ — список слов
    return: словарь {doc_id: {слово: tfidf_значение}}
    """
    d = defaultdict(dict)
    m1 = self._idf(documents)
    for idx, arr in enumerate(documents):
        m2 = self._tf(arr)
        for word in arr:
            d[idx][word] = m1[word] * m2[word]

    return d
        

Tfidfindex.cocompute_tfidf = compute_tfidf
```
